# What is matplotlib ?
Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. It was introduced by John Hunter in the year 2002.

One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.

## Basics of Matplotlib

### Importing the matplotlib module

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Basic Graph

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.show()
```

### plot() basically plot x vs y graph

### Applying the title to the graph

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph")
plt.show()
```

### Setting the x and y label

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph")
plt.xlabel("x")
plt.ylabel("Y")
plt.show()
```

### Changing the font of the title

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph",fontdict={'fontname':"Consolas","fontsize":30}).set(color="blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Specifying the title of the graph to center,left or right 
Using loc parameter

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph",fontdict={'fontname':"Consolas","fontsize":30},loc="left").set(color="blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Here you can see that y label is rotated so to do that
use rotation parameters of text class

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph",fontdict={'fontname':"Consolas","fontsize":30},loc="left").set(color="blue")
plt.xlabel("X")
plt.ylabel("Y",rotation="horizontal")
plt.show()
```

### We can also change the font of the x and y label

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph",fontdict={'fontname':"Consolas","fontsize":30},loc="left").set(color="blue")
plt.xlabel("X")
plt.ylabel("Y",rotation="horizontal",fontsize=20,position=(0,50))
plt.show()
```

### Setting the xticks and yticks means changing the number having on x axis and y axis

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph",fontdict={'fontname':"Consolas","fontsize":30},loc="left").set(color="blue")
plt.xlabel("X")
plt.ylabel("Y",rotation="horizontal")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,10,20,30,40,50])
plt.show()
```

### Setting the label on the xticks
you can also apply the label on yticks
but while applying ticks and label should be there

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph",fontdict={'fontname':"Consolas","fontsize":30},loc="left").set(color="blue")
plt.xlabel("X")
plt.ylabel("Y",rotation="horizontal")
plt.xticks([0,1,2,3,4,5],labels=["Zero",'First',"Second","Third","Fourth","Fifth"])
plt.yticks([0,10,20,30,40,50])
plt.show()
```

### Applying the text properties on the xticks and yticks

```python
x=[1,2,3,4]
y=[10,11,50,1]
plt.plot(x,y)
plt.title("X / Y Graph",fontdict={'fontname':"Consolas","fontsize":30},loc="left").set(color="blue")
plt.xlabel("X")
plt.ylabel("Y",rotation="horizontal")
plt.xticks([0,1,2,3,4,5],labels=["Zero",'First',"Second","Third","Fourth","Fifth"],color="Red")  ## change to red color
plt.yticks([0,10,20,30,40,50])
plt.show()
```

## Labeling the plot

##### Using legend() method to show the labels on the graph

```python
x1=np.linspace(1,50,8,dtype=int)
y1=np.linspace(20,70,8,dtype=int)
x2=np.linspace(1,100,8,dtype=int)
y2=np.linspace(50,100,8,dtype=int)
plt.plot(x1,y1,label="x1 vs y1")
plt.plot(x2,y2,label="x2 vs y2")
plt.legend()  
plt.show()
```

### Setting the color of the line in the graph
using color parameter of the text 

There are abbrevation for the color you can either use the hex color also

```python
x1=np.linspace(1,50,8,dtype=int)
y1=np.linspace(20,70,8,dtype=int)
x2=np.linspace(1,100,8,dtype=int)
y2=np.linspace(50,100,8,dtype=int)
plt.plot(x1,y1,label="x1 vs y1",color="red")
plt.plot(x2,y2,label="x2 vs y2",color="green")
plt.legend()  
plt.show()
```

### changing the linewidth of the graph

```python
x1=np.linspace(1,50,8,dtype=int)
y1=np.linspace(20,70,8,dtype=int)
x2=np.linspace(1,100,8,dtype=int)
y2=np.linspace(50,100,8,dtype=int)
plt.plot(x1,y1,label="x1 vs y1",color="#f403fc",linewidth=6) # applying the hex color
plt.plot(x2,y2,label="x2 vs y2")
plt.legend()  
plt.show()
```

### If you want to add marker means where the point is plotted

```python
plt.plot(x1,y1,label="x1 vs y1",color="#f403fc",linewidth=6) # applying the hex color
plt.plot(x2,y2,label="x2 vs y2",marker="o") # applying the marker
plt.legend()  
plt.show()
```

### Applying the markersize of the line

```python
plt.plot(x1,y1,label="x1 vs y1",color="#f403fc",linewidth=6) # applying the hex color
plt.plot(x2,y2,label="x2 vs y2",marker="o",markersize=10) # applying the marker
plt.legend()  
plt.show()
```

### changing the markeredgecolor

```python
plt.plot(x1,y1,label="x1 vs y1",color="#f403fc",linewidth=6) # applying the hex color
plt.plot(x2,y2,label="x2 vs y2",marker="o",markeredgecolor="r") # applying the marker
plt.legend()  
plt.show()
```

### Changing the markerface color

```python
plt.plot(x1,y1,label="x1 vs y1",color="#f403fc",linewidth=6) # applying the hex color
plt.plot(x2,y2,label="x2 vs y2",marker="o",markerfacecolor="black") # applying the marker
plt.legend(title="Graph") # applying the title to the legend 
plt.show()
```

### Changing the line style in the graph </br>
'-'

solid line style

'--'

dashed line style

'-.'

dash-dot line style

':'

dotted line style

```python
plt.plot(x1,y1,label="x1 vs y1",color="#f403fc",linewidth=2,linestyle="--") # applying the hex color
plt.plot(x2,y2,label="x2 vs y2",marker="o",markerfacecolor="black",linestyle=":") # applying the marker
plt.legend(title="Graph") # applying the title to the legend 
plt.show()
```

### You can also use the format string

### Resizing the graph
using figure(figsize=(float,float))

also applying the figure text using figtext(x,y,string,fontdict-{})

```python
plt.figure(figsize=(2,3))
plt.figtext(1,0,s="1st Figure")
plt.plot(x1,y1,'b^--',label="x1 vs y1") # applying the hex color
plt.plot(x2,y2,label="x2 vs y2",marker="o",markerfacecolor="black") # applying the marker
plt.legend(title="Graph") # applying the title to the legend 
plt.show()
```

## bar plot

Plotting the bar graph of city and their there average ordered quantity

```python
citys=df['city'].unique()
citys
```

```python
df.loc[df["city"]==citys[0]]["quantityordered"].mean()
```

```python
def getmeans():
    meanslist=[]
    for i in citys:
        meanslist.append(df.loc[df["city"]==i]["quantityordered"].mean())
    return meanslist
        
means=getmeans()
means
```

```python
plt.figure(figsize=(5,5))
plt.bar(x=citys,height=means,color=["red","green","yellow"],edgecolor="black",label=citys,hatch=["/","*","|"])
plt.xlabel("City")
plt.ylabel("Means of Order")
plt.minorticks_on()
plt.title("Average Quantity Order form Each City")
plt.legend(loc=(1.02,0.80),title="Cities")
plt.show()
```

## Subplots

subplot(row,column,index)

```python
plt.subplot(2,2,1)
```

```python
def calper(x,sum):
    return x/sum*100
```

```python


y2003=df.loc[df["year_id"]==2003]["sales"]
y2003=y2003.apply(calper,args=(y2003.sum(),))

y2004=df.loc[df["year_id"]==2004]["sales"]
y2004=y2004.apply(calper,args=(y2004.sum(),))

y2005=df.loc[df["year_id"]==2005]["sales"]
y2005=y2005.apply(calper,args=(y2005.sum(),))
```

```python
plt.suptitle("Sales ON Years")
plt.subplot(2,2,1,label="2003")
plt.plot(y2003)
plt.title("2003")

plt.subplot(2,2,2)
plt.plot(y2004)
plt.title("2004")

plt.subplot(2,2,3)
plt.plot(y2005)
plt.title("2005")
plt.show()
```

## Annotate in matplotlib

Consider the Plot 

In this we are going to add the text and write some mathematical equations

```python
plt.plot(y2005,label="2005",marker=".",linestyle="-.")
plt.xlabel("")
plt.title("2005")
plt.text(x=130,y=4,s="Sales",fontsize=15,fontfamily="Consolas")  ## Adds the text in the graph
plt.legend()
plt.show()
```

## Using Annotation

```python
plt.plot(y2005,label="2005",marker=".",linestyle="-.")
plt.title("2005")
plt.annotate("HERE",xy=(128,7.6))
plt.legend()
plt.show()
```

### Adding the arrow 

```python
plt.plot(y2005,label="2005",marker=".",linestyle="-.")
plt.title("2005")
plt.annotate(" ",xy=(128,7.6),arrowprops={"arrowstyle":"->","relpos":(17,50)})
plt.legend()
plt.show()
```

### Writing Mathematical equation

Math text should be started and ended with `$` Sign only.
It should be enclosed in `r"$ $"`




```python
plt.plot(y2005,label="2005",marker=".",linestyle="-.")
plt.title("2005")
plt.annotate(r"$\alpha$",xy=(128,7.6))
plt.legend()
plt.show()
```
