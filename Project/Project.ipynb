#### Changing the datatype of `Last Updated` Column to Datatime Datatype

```python
df["Last Updated"]=pd.to_datetime(df["Last Updated"])
```

#### Changing the values of Price 

You can see that there is $ sign in front of every price so replace it with the blank

```python
df.value_counts(subset="Price")
```

```python
def changedoll(row):
    dl=row.replace("$","")
    return dl
    
    
df["Price"]=df["Price"].apply(changedoll)
```

```python
df["Price"].value_counts().head(50)
```

Same Goes With Installs

```python
df["Installs"].value_counts()
```

```python
def instl(row):
    commspl=row.replace(",","")
    return commspl.replace("+","")

df["Installs"]=df["Installs"].apply(instl)
```

```python
df["Installs"].value_counts()
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df=df.convert_dtypes()
```

```python
df.drop(columns="index",inplace=True)
```

### Changing the datatypes of `Reviews`, `Installs` and `Price` to float

```python
df=df.astype({"Reviews":float,"Installs":float,"Price":float})
```

```python
df.dtypes
```

#### Plotting the Count Plot To show the Content of Apps mostly are

```python
plt.figure(figsize=(15,15))
sns.countplot(data=df,x="Content Rating",hue="Type")
plt.title("Content Of Apps")
plt.show()
```

```python
df
```

### Types of Apps mostly available

```python
free=df.loc[df["Type"]=="Free"]["Type"].count()
paid=df.loc[df["Type"]=="Paid"]["Type"].count()
plt.pie([free,paid],labels=["Free","Paid"],autopct="%.2f %%",explode=[0,0.2],shadow=True,colors=["#cccc00","pink"])
plt.legend(title="Content")
plt.show()
```

```python
df
```

```python
app_by_category=df.groupby(by="Category")["App"].count()
app_by_category.sort_values(inplace=True)
app_by_category
```

### Which Category have most apps

```python
plt.figure(figsize=(15,15))
sns.barplot(x=app_by_category,y=app_by_category.index,palette="Blues")
plt.xlabel("Apps Count")
plt.title("Most Apps By Category")
plt.show()
```

```python
df
```

### Top 5 Apps in Rating and Reviews

```python
top5=df.nlargest(n=5,columns=["Rating","Reviews"])
top5
```

```python
plt.figure(figsize=(15,15))
sns.barplot(data=top5,x="App",y="Reviews",width=0.4)
plt.title("Top 5 Apps")
plt.show()
```

### Top 5 Paid Apps With Reviews

```python
top5_paid=df.loc[df["Type"]=="Paid"]
```

```python
top5_paid=top5_paid.nlargest(n=5,columns="Price")
top5_paid
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=top5_paid,y="App",x="Reviews",width=0.2,label=top5_paid.App)
plt.legend()
plt.title("Top 5 Paid Apps Reviews")
plt.show()
```
