# What is Seaborn ?
Seaborn is a library mostly used for statistical plotting in Python. It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive.

### Importing the required modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
```

### HUE Parameter is used to order the data in category wise

### Loading the data set

```python
df=pd.read_csv("sales_data_sample.csv",encoding="Latin-1")
df.rename(str.lower,axis=1,inplace=True)
df
```

### Lineplot

```python
sns.lineplot(data=df)

plt.show()
```

Shows the overall line with default legend

```python
df
```

```python
sns.lineplot(data=df,x="quantityordered",y="msrp",hue="year_id",markers=["o","^","*"],sort=True,err_style="band") # See the style
plt.show()
```

```python
sns.lineplot(data=df,x="quantityordered",y="msrp",hue="year_id",markers=["o","^","*"],sort=True,err_style="bars") # See the style
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=df,x="quantityordered",y="msrp",markers=["o","^","*"],sort=True,err_kws={"visible":False}) # See the style
sns.lineplot(data=df,x="quantityordered",y="orderlinenumber",markers=["o","^","*"],sort=True,err_kws={"visible":False})
plt.show()
```

The Errors bars are removed

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=df,x="quantityordered",y="msrp",hue="year_id",markers=["o","^","*"],sort=True,err_style="band",err_kws={"visible":False},palette=["blue","green","red"],legend="full") # See the style
plt.show()
```

## Bar Plot

```python
df=sns.load_dataset("titanic")
# df.drop(columns="Unnamed: 0",inplace=True)
df
```

```python
sns.barplot(data=df,x="class",y="age")
plt.show()
```

## Jointplot

```python
ndf=sns.load_dataset('iris')
ndf
```

```python
sns.jointplot(data=ndf,x="sepal_length",y="petal_length")
plt.show()
```

```python
sns.jointplot(data=ndf,x="sepal_length",y="petal_length",kind="hex")
plt.show()
```

## KDE Plot

```python
sns.kdeplot(data=ndf,x="sepal_width",y="petal_width",hue="species")
plt.grid()
plt.show()
```

## Pair Plot

```python
sns.pairplot(data=ndf,dropna=True,markers="o",kind='kde')
plt.show()
```

## Rugplot

```python
sns.rugplot(data=ndf,x="sepal_length",y="sepal_width")
sns.kdeplot(data=ndf,x="sepal_length",y="sepal_width")
plt.show()
```

Here we combine the many plot in one plot

## Styling

```python
sns.set_style(style="dark")
sns.kdeplot(data=ndf,x="sepal_length",y="sepal_width")
plt.show()
```

## Setting the theme

```python
sns.set_style(style="white")
sns.set_theme(context={'font.size': 15.0},style={'axes.labelcolor': 'red'})
sns.kdeplot(data=ndf,x="sepal_length",y="sepal_width")
plt.show()
```

## Setting the axes 

```python
# sns.set_style(style="white")
sns.despine(left=True,top=False)
sns.displot(data=df,x="age",hue="sex",kind="kde")
plt.show()
```

# Categorical Plots

## Count Plot

```python
sns.countplot(data=df,x="who",hatch=["/","o","*"],width=0.5)
plt.show()
```

## Box Plot

```python
sns.boxplot(data=ndf)
plt.show()
```

## Violin Plot

```python
sns.violinplot(data=ndf)
plt.show()
```

## Stripplot

```python
sns.stripplot(data=df,x="age",y="class",hue="sex",orient="h",dodge=True,jitter=True)
plt.legend(loc=(0,1.05),title="SEX",labels=["Male","Female"])
plt.show()
```

### Using Palette

```python
sns.stripplot(data=df,x="age",y="class",hue="sex",orient="h",dodge=True,jitter=True,palette="Spectral",edgecolor="black")
plt.legend(loc=(0,1.05),title="SEX",labels=["Male","Female"])
plt.show()
```

# Matrix Plots

## Heatmaps

### Presenting the data in the matrix form

```python
matdf=df.corr()
matdf
```

```python
sns.set_context(context={"font.size":10})
sns.heatmap(matdf,cmap="Blues",annot=True)
plt.show()
```

## ClusterMaps

```python
cdf=df.pivot_table(values="fare",index="age",columns="class").dropna()
sns.clustermap(data=cdf)
plt.show()
```

# Regression Plot

```python
df
```

```python
sns.lmplot(data=df,y="fare",x="age",hue="sex",col="who",row="class")
plt.show()
```

## FacetGrid

```python
gg=sns.FacetGrid(data=df,row="pclass",col="alive",hue="sex")
gg.map(sns.kdeplot,"age")

plt.show()
```
